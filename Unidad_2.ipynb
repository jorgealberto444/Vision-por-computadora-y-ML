```python
# Importación de modelos de Machine Learning desde scikit-learn

# LogisticRegression representa un modelo lineal clásico utilizado
# para problemas de clasificación binaria. Este modelo realiza una
# combinación lineal de las variables de entrada para generar una predicción.

from sklearn.linear_model import LogisticRegression

# MLPClassifier implementa un perceptrón multicapa (Multi Layer Perceptron),
# que es una red neuronal artificial compuesta por capas ocultas.
# Este tipo de modelo permite aprender relaciones no lineales entre los datos.

from sklearn.neural_network import MLPClassifier

# Se imprime la estructura del modelo de regresión logística
print(LogisticRegression)

# Se crea un modelo MLP con dos capas ocultas de 64 neuronas cada una
# hidden_layer_sizes indica el número de neuronas en cada capa oculta
print(MLPClassifier(hidden_layer_sizes=(64,64)))
```

```python
# Importación de la biblioteca NumPy para operaciones matemáticas
import numpy as np

# Vector de entrada que representa las características del dato
x = np.array([1.0, 2.0])

# Matriz de pesos de la capa neuronal
# Cada fila representa los pesos asociados a una neurona
w = np.array([[0.5, -1.0],
              [1.0, 0.5]])

# Vector de sesgos (bias)
# Permite desplazar la función de activación
b = np.array([0.1, -0.2])

# Operación fundamental de una capa neuronal
# y = Wx + b
# dot() realiza el producto matricial
y = np.dot(w, x) + b

# Se imprime la salida de la capa
print(y)
```

```python
import numpy as np

# Definición de la función de activación ReLU
# Esta función reemplaza los valores negativos por cero
def relu(x):
    return np.maximum(0, x)

# Vector de entrada (features)
x = np.array([1.0, 2.0, -1.0])

# Pesos de la neurona
w = np.array([0.5, -0.3, 0.8])

# Sesgo o bias
b = 0.1

# Cálculo de la salida lineal de la neurona
# z = wx + b
z = np.dot(w, x) + b

# Simulación del paso de los datos por varias capas
for i in range(3):
    # np.eye genera una matriz identidad que mantiene el tamaño del vector
    x = relu(np.dot(np.eye(3), x))
    print(f"capa {i+1}: {x}")

# Resultado antes de aplicar activación
print("salida lineal: ", z)
```

```python
import numpy as np

# Función ReLU
def relu(x):
    return np.maximum(0, x)

# Vector de entrada
x = np.array([1.0, -1.0])

# Simulación de múltiples capas
for i in range(3):

    # np.eye genera una matriz identidad
    # Se utiliza para mantener la dimensión del vector
    x = relu(np.dot(np.eye(2), x))

    print(f"capa {i+1}: {x}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Creación de una red neuronal multicapa (MLP)
model = Sequential([

    # Capa oculta con 16 neuronas
    Dense(16, activation='relu', input_shape=(10,)),

    # Segunda capa oculta con 8 neuronas
    Dense(8, activation='relu'),

    # Capa de salida para clasificación binaria
    Dense(1, activation='sigmoid')

])

# Mostrar estructura del modelo
model.summary()
```

```python
from sklearn.datasets import make_classification
from sklearn.neural_network import MLPClassifier

# Generación de un conjunto de datos artificial
X, y = make_classification(n_samples=200, n_features=4)

# Creación de un modelo MLP
clf = MLPClassifier(hidden_layer_sizes=(10,10), max_iter=500)

# Entrenamiento del modelo
clf.fit(X, y)

# Evaluación del modelo
print("precision:", clf.score(X, y))
```

```python
import numpy as np

# Derivada de la función f(x)=x²
def grad(x):
    return 2 * x

# Valor inicial
x = 10.0

# Tasa de aprendizaje
lr = 0.1

# Proceso iterativo de optimización
for i in range(10):

    # Actualización del parámetro
    x = x - lr * grad(x)

    print(f"iteracion {i+1}: x = {x}")
```

```python
import tensorflow as tf

# Variable entrenable
x = tf.Variable(3.0)

# Cálculo automático del gradiente
with tf.GradientTape() as tape:

    # Función matemática
    y = x**2

# Derivada de y con respecto a x
dy_dx = tape.gradient(y, x)

print(dy_dx.numpy())
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

# Valores reales
y_true = [0, 1, 1, 0, 1]

# Predicciones del modelo
y_pred = [0, 1, 0, 0, 1]

# Cálculo de la exactitud
print("Accuracy:", accuracy_score(y_true, y_pred))

# Matriz de confusión
print("Matriz de confusion:\n", confusion_matrix(y_true, y_pred))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Creación de un modelo MLP
model = Sequential([

    Dense(32, activation='relu', input_shape=(10,)),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')

])

# Compilación del modelo
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Visualización del modelo
model.summary()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generación de valores para el eje x
x = np.linspace(-5, 5, 100)

# Cálculo de diferentes funciones de activación
relu = np.maximum(0, x)
sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)

# Graficar las funciones
plt.plot(x, relu, label="ReLU")
plt.plot(x, sigmoid, label="Sigmoid")
plt.plot(x, tanh, label="Tanh")

# Configuración de la gráfica
plt.legend()
plt.grid(True)

# Mostrar gráfica
plt.show()
```
